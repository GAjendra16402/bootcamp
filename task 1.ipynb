```python
import pandas as pd

def generate_car_matrix(dataset_path):
    
    df = pd.read_csv("C:\\Users\\gajendra singh\\OneDrive\\Desktop\\pandas\\datasets\\dataset-1.csv")

   
    pivot_df = df.pivot(index='id_1', columns='id_2', values='car')

   
    pivot_df = pivot_df.fillna(0)

    
    pivot_df.values[[range(len(pivot_df))]*2] = 0

    return pivot_df

result = generate_car_matrix(dataset_path)
print(result)
```

# create a python function named get_type_count that takes the dataset-1.csv as
a DataFrame. Add a new categorical column car_type based on values of the column car:
    
    
    low for values less than or equal to 15
    
    medium for values greater than 15 and less than or equal to 25
    
    high for values greater than 25
    
    calculate the count of occurence for each car_type category and return the 
    result as a dictionary. sort the dicitonary alphabetically based on keys.

```python
def get_type_count(df):
    # Define the conditions for the car_type categories
    conditions = [
        (df['car'] <= 15),
        (df['car'] > 15) & (df['car'] <= 25),
        (df['car'] > 25)
    ]

    # Define the corresponding car_type values
    car_types = ['low', 'medium', 'high']

    # Create the 'car_type' column based on the conditions
    df['car_type'] = np.select(conditions, car_types, default=np.nan)

    print("Modified DataFrame:")
    print(df)

    # Calculate the count of occurrences for each 'car_type' category
    type_counts = df['car_type'].value_counts().to_dict()

    # Sort the dictionary alphabetically based on keys
    sorted_type_counts = {k: type_counts[k] for k in sorted(type_counts)}

    return type_counts

# Example usage
dataset_path = "C:\\Users\\gajendra singh\\OneDrive\\Desktop\\pandas\\datasets\\dataset-1.csv"  # Replace with the actual path to your CSV file
df = pd.read_csv(dataset_path)
result = get_type_count(df)
print(result)
```
