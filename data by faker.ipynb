```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Define the number of rows and columns
num_rows = 50
num_columns = 10

# Create an empty DataFrame with column names
columns = ['Student_ID', 'First_Name', 'Last_Name', 'Date_of_Birth', 'Gender', 'Grade_Level',
           'Address', 'City', 'State', 'ZIP_Code']
data = pd.DataFrame(columns=columns)

# Populate the DataFrame with more meaningful data
for i in range(num_rows):
    row_data = [i + 1,
                fake.first_name(),
                fake.last_name(),
                fake.date_of_birth(minimum_age=5, maximum_age=18).strftime('%Y-%m-%d'),
                fake.random_element(["Male", "Female"]),
                fake.random_int(9, 12),
                fake.street_address(),
                fake.city(),
                fake.state_abbr(),
                fake.zipcode()]
    
    data.loc[i] = row_data

# Save the DataFrame to a CSV file with column names
data.to_csv('school_information_data_10x50.csv', index=False)
```

```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Define the number of rows and columns
num_rows = 50
num_columns = 10

# Create an empty DataFrame with column names
columns = ['Employee_ID', 'First_Name', 'Last_Name', 'Date_of_Birth', 'Gender', 'Department',
           'Position', 'Office_Location', 'Email', 'Phone_Number']
office_data = pd.DataFrame(columns=columns)

# Populate the DataFrame with more meaningful office-related data
for i in range(num_rows):
    row_data = [i + 1,
                fake.first_name(),
                fake.last_name(),
                fake.date_of_birth(minimum_age=22, maximum_age=60).strftime('%Y-%m-%d'),
                fake.random_element(["Male", "Female"]),
                fake.random_element(["Finance", "Marketing", "IT", "Human Resources"]),
                fake.random_element(["Manager", "Analyst", "Developer", "Coordinator"]),
                fake.random_element(["New York", "London", "Tokyo", "Sydney"]),
                fake.email(),
                fake.phone_number()]
    
    office_data.loc[i] = row_data

# Save the DataFrame to a CSV file with column names
office_data.to_csv('office_information_data_10x50.csv', index=False)
```
